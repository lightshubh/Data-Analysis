```python
df_cleaning= data.copy()
```

```python
# checking the non null values in "Unnamed : 10" column
df_cleaning[~df_cleaning['Unnamed: 10'].isna()]
```

```python
# checking the unique values from 'Unnamed: 10' column
df_cleaning['Unnamed: 10'].unique()
```

```python
# "Unnamed: 9" is null column and Unnamed: 10" has all the irrelevent values.
# hence dropping these columns

df_cleaning.drop(columns= ['Unnamed: 9', 'Unnamed: 10'], inplace= True)
```

```python
df_cleaning.dropna(axis= 0, how= 'all', inplace= True)
```

```python
df_cleaning.head(10)
```

```python
# creating a new list for iteration
column_values= []
counter= 0
for i in df_cleaning['Row Type']:
    if 'first name' in i:
        counter += 1
    column_values.append(counter)    
```

```python
# adding the iteration column in the table
df_cleaning['iteration']= column_values
```

```python
df_cleaning.head(10)
```

```python
# dropping extra rows:

df_cleaning= df_cleaning[df_cleaning['Row Type'] != 'Row Type']
```

```python
df_cleaning.head(10)
```

```python
df_cleaning_1= df_cleaning[df_cleaning['Row Type'].str.contains('first name')]
df_cleaning_1
```

```python
# droping extra column from 'df_cleaning_1' column
df_cleaning_1= df_cleaning_1.drop(columns= ['Speed1', 'Speed2', 'Electricity', 'Effort', 'Weight', 'Torque'])
```

```python
# renaming the columns:
df_cleaning_1.rename(columns= {'Row Type': 'first name',
                               'Iter Number' : 'last name',
                               'Power1' : 'date'}, inplace= True)
```

```python
# removing extra values:

df_cleaning_1['first name']= df_cleaning_1['first name'].str.lstrip('first name: ')
df_cleaning_1['last name']= df_cleaning_1['last name'].str.lstrip('last name: ')
df_cleaning_1['date']= df_cleaning_1['date'].str.lstrip('date:')
```

```python
df_cleaning_1
```

```python
df_cleaning_2= df_cleaning[~df_cleaning['Row Type'].str.contains('first name')]
```

```python
# mergin both datasets:
df_cleaning_3= pd.merge(left= df_cleaning_1, right= df_cleaning_2, how= 'inner', on= 'iteration')
```

```python
df_cleaning_3
```

```python
# further cleaning the table

df_cleaning_4= df_cleaning_3[~((df_cleaning_3['Row Type'] =='Average') | (df_cleaning_3['Row Type'] == 'Maximum') |
               (df_cleaning_3['Row Type'] =='Std.Dev.') | (df_cleaning_3['Row Type'] == 'Total'))]
```

```python
df_cleaning_4.drop(columns= 'iteration', inplace= True)
```

```python
df_cleaning_4.reset_index(drop= True, inplace= True)
```
