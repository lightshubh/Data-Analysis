```python
df_cars['Make'].fillna(value='Toyota', axis=0, inplace= True)
df_cars['Model'].fillna(value='G35 4dr', axis=0, inplace= True)
df_cars['Type'].fillna(value='Sedan', axis=0, inplace= True)
df_cars['Origin'].fillna(value='Asia', axis=0, inplace= True)
df_cars['DriveTrain'].fillna(value='Front', axis=0, inplace= True)
df_cars['MSRP'].fillna(value='$21,055', axis=0, inplace= True)
df_cars['Invoice'].fillna(value='$68,306', axis=0, inplace= True)
```

```python
counter= 7
while(counter <= 14):
    df_cars.iloc[:, counter].fillna(df_cars.iloc[:, counter].mean(), axis=0, inplace= True)
    counter += 1
```
