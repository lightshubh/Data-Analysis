```python
data= pd.read_excel(r"C:\Users\light\Desktop\Data - Survey Monkey-Edited.xlsx", sheet_name= 'Edited_Data')
```

```python
df_monkey= data.copy()
```

```python
# dropping the unwanted columns:
df_monkey.drop(columns= ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1'], inplace= True)
```

```python
df_monkey
```

```python
column_list= list(df_monkey.columns)
column_list
```

```python
vars= column_list[:8]
vars
```

```python
val= column_list[8:]
val
```

```python
# melting the data so we have more rows then columns:

df_monkey_melted= df_monkey.melt(id_vars= vars, value_vars= val, var_name= 'Questions_&_Subquestions', value_name='Answers')
```

```python
df_monkey_melted
```

```python
# renaming some of the columns:

df_monkey_melted.rename(columns= {'Identify which division you work in. - Response' : 'Division', 
                                  'Identify which division you work in. - Other (please specify)' : 'Division(Other)', 
                                  'Which of the following best describes your position level? - Response' : 'Position_Level', 
                                  'Which generation are you apart of? - Response' : 'Generation', 
                                  'Please select the gender in which you identify. - Response' : 'Gender',
                                  'Which duration range best aligns with your tenure at your company? - Response' : 'Tenure',
                                  'Which of the following best describes your employment type? - Response' : 'Employment_Type'},
                       inplace= True)
```

```python
df_monkey_melted
```

```python
# importing the Questions sheet
df_questions= pd.read_excel(r"C:\Users\light\Desktop\Data - Survey Monkey-Edited.xlsx", sheet_name= 'Questions')
```

```python
df_questions
```

```python
# dropping unwanted columns from df_questions

df_questions.drop(columns= ['Raw_Questions', 'Raw_Subquestions', 'Subquestion'], inplace= True)
```

```python
df_questions
```

```python
# Merging this columns to the 'df_monkey_melted' table:

df_monkey_melted_1= pd.merge(left= df_monkey_melted, right= df_questions, left_on= 'Questions_&_Subquestions', right_on='Questions_&_Subquestions', how= 'left')

```

```python
# creating checkpoint
df_monkey_Edited= df_monkey_melted_1
df_monkey_Edited
```

```python
# renaming some columns:

df_monkey_Edited.rename(columns= {'Identify which division you work in.-Response' : 'Division',
                                  'Identify which division you work in.-Other (please specify)' : 'Division(Other)',
                                  'Which of the following best describes your position level?-Response': 'Position_Level',
                                  'Which generation are you apart of?-Response' : 'Generation',
                                  'Please select the gender in which you identify.-Response' : 'Gender',
                                  'Which duration range best aligns with your tenure at your company?-Response' : 'Tenure',
                                  'Which of the following best describes your employment type?-Response' : 'Employment_Type'},
                       inplace= True)
```

```python
df_monkey_Edited
```

```python
# rearranging some columns:
df_monkey_Edited= df_monkey_Edited.loc[:, ['Respondent ID', 'Division', 'Division(Other)', 'Position_Level', 
                 'Generation', 'Gender', 'Tenure', 'Employment_Type', 'Questions', 'Questions_&_Subquestions', 'Answers']]
```

```python
df_monkey_Edited
```

```python
df_monkey_Edited.shape
```

```python
# Calculating questions answerd by number of respondent:

df_answered= df_monkey_Edited[df_monkey_Edited['Answers'].notna()]
```

```python
df_answered
```

```python
df_answered.shape
```

```python
df_answered= df_answered.groupby('Questions')['Respondent ID'].nunique().reset_index()
df_answered.rename(columns= {'Respondent ID' : 'Total_Respondent'}, inplace= True)
df_answered
```

```python
# adding this column to 'df_monkey_Edited' table

df_monkey_Edited_1= pd.merge(left= df_monkey_Edited, right= df_answered, how= 'left', left_on= 'Questions', right_on= 'Questions')
```

```python
df_monkey_Edited_1
```

```python
# calculating how many respondent gave the same answers:

df_same_ans= df_monkey_Edited_1.groupby(['Questions_&_Subquestions', 'Answers'])['Respondent ID'].nunique().reset_index()
df_same_ans.rename(columns= {'Respondent ID' : 'Same_Answers'}, inplace= True)
df_same_ans
```

```python
#  Merging the 'df_same_ans' column in 'df_monkey_Edited_1' column:

df_final_table= pd.merge(left= df_monkey_Edited_1, right= df_same_ans, how= 'left', left_on= ['Questions_&_Subquestions', 'Answers'], right_on= ['Questions_&_Subquestions', 'Answers'])
```

```python
df_final_table
```

```python
# exporting this table:
df_final_table.to_excel('Filtered Monkey Data.xlsx', index= False)
```
