```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

dataset link: https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results

```python
df_athlete= pd.read_csv(r"C:\Users\light\Desktop\data analysis\datasets\120 years of Olympic history athletes and results\athlete_events.csv")
df_region= pd.read_csv(r"C:\Users\light\Desktop\data analysis\datasets\120 years of Olympic history athletes and results\noc_regions.csv")
```

```python
# calculating the same result for 'summer' season:

female_athlete_summer= df_athlete_merged[(df_athlete_merged['Season'] == 'Summer') & (df_athlete_merged['Sex'] == 'F')].groupby('Year')['ID'].count().sort_index()                 
female_athlete_summer.head()
```

```python
female_athlete_summer.plot()
plt.show()
```

```python
# gold medals

gold_medals= df_athlete_merged[df_athlete_merged['Medal'] == 'Gold']
gold_medals.head()
```

```python
# taking the values which are different from NaNs

gold_medals_1= gold_medals[~gold_medals['Age'].isna()]
gold_medals_1.head()
```

```python
# 60 years and higher aged athlete who won gold 

gold_medals_60= gold_medals_1[gold_medals_1['Age'] >= 60]
gold_medals_60
```

```python
# visualing the result:

plt.tight_layout()
sns.countplot(x= gold_medals_60['Sport'])
plt.title('Gold Medal for Athlete(60 Years and Above)', weight= 'bold')

plt.show()
```

```python
# total gold medal for each country:

total_gold= df_athlete_merged.query('Medal == "Gold"')['Region'].value_counts().reset_index(name= 'Medal')
total_gold.head(5)
```

```python
# ploting the above result:
total_gold_10= total_gold.head(10)

sns.catplot(x= 'index', y= 'Medal', data= total_gold_10, kind= 'bar')
plt.xticks(rotation= 'vertical')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.title('Top 10 Countries with Most Gold Medals')
plt.show()
```

```python
# data of 'Rio' olympic(2016)

df_2016= df_athlete_merged[(df_athlete_merged['Year'] == 2016) & (df_athlete_merged['City'] == 'Rio de Janeiro')]
df_2016
```

```python
# in 2016 who won gold medals:
df_2016.query('Medal == "Gold"').head()
```

```python
df_2016_gold= df_2016.query('Medal == "Gold"')['Region'].value_counts().reset_index(name= 'Gold_Medal')
df_2016_gold.head()
```

```python
# visualising above results:
df_2016_gold_20= df_2016_gold.head(20)
plt.barh(df_2016_gold_20['index'], df_2016_gold_20['Gold_Medal'])
plt.xlabel('Number of Medals')
plt.ylabel('Countries')
plt.title('Top 20 Countries with Most Gold Medals(Rio 2016)', weight= 'bold')
sns.despine()
plt.show()
```

```python
# scatter plot for height of male and females who won medals(any)
df_2016_medal= df_athlete_merged[(df_athlete_merged['Medal'].notnull()) & (df_athlete_merged['Height'].notnull())]
```

```python
plt.figure(figsize= (10, 6))
sns.scatterplot(x= 'Weight', y= 'Height',data= df_2016_medal, hue= 'Sex')
plt.title('Male and Female Athlete(2016 Rio)')
plt.show()
```
