```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

Taking the data from:  https://censusindia.gov.in/Tables_Published/A-Series/A-Series_links/t_00_003.aspx

```python
data= pd.read_html('https://censusindia.gov.in/Tables_Published/A-Series/A-Series_links/t_00_003.aspx',
                  header= 1)
```

```python
df_india= data[2].copy()
df_india
```

```python
# cleaning / rearranging the data: 

df_india_1= df_india.iloc[1:36, 1:]
df_india_1
```

```python
# renaming the columns:

df_india_1.rename(columns= {'Persons' : 'Total_Population',
                            'Sex ratio' : 'Sex_Ratio'}, inplace= True)
```

```python
# removing the '@' sign from 'Manipur':

# df_india_1['State / UT'].apply(lambda x : 'Manipur' if 'Manipur@' else x)

df_india_1['State / UT']= ['Manipur' if i == 'Manipur@' else i for i in df_india_1['State / UT']]
```

```python
df_india_1

# i think we have cleaned the dataset:
```

### Note:
If we want we can also drop 'Total_Population' and  'Sex ratio' columns, because we can calculate these terms with the help of male and female population. In short we have only three main columns: ['State / UT', 'Males', 'Females'] 

```python
# exploring the data set

df_india_1.dtypes
```

```python
# changing the data type for some columns:

def converting(list= []):
    for i in list:
        df_india_1[i]= df_india_1[i].astype(int)
        
converting(['Total_Population', 'Males', 'Females', 'Sex  ratio'])        
```

```python
df_india_1.dtypes
```

### Top 10

```python
# top 10 and least 10 'state / ut' in terms of population
# top 10
top_10_pop= df_india_1[['State / UT', 'Total_Population']].sort_values(by= 'Total_Population', ascending= False)
top_10_pop.head(10)
```

```python
# least 10
least_10_pop= df_india_1[['State / UT', 'Total_Population']].sort_values(by= 'Total_Population')
least_10_pop.head(10)
```

```python
# visualising both of the results

fig, (ax1, ax2)= plt.subplots(1, 2, figsize= (10, 4))
ax1.bar(x= top_10_pop.head(10)['State / UT'], height= top_10_pop.head(10)['Total_Population'])
ax1.set_xticks(top_10_pop.head(10)['State / UT'])
ax1.set_title('Top 10 State(Population)', weight= 'bold')
ax1.set_xticklabels(labels= top_10_pop.head(10)['State / UT'], rotation= 'vertical')


ax2.bar(x= least_10_pop.head(10)['State / UT'], height= least_10_pop.head(10)['Total_Population'])
ax2.set_xticks(least_10_pop.head(10)['State / UT'])
ax2.set_title('Least 10 State(Population)', weight= 'bold')
ax2.set_xticklabels(labels= least_10_pop.head(10)['State / UT'], rotation= 'vertical')

plt.show()
```

```python
# comparison of female population where male population is high

df_com_max= df_india_1[['State / UT', 'Males', 'Females']].sort_values(by= 'Males', ascending= False)
df_com_max.head()
```

```python
# for top 20 State / UT

df_com_max.set_index('State / UT').head(20).plot.bar(figsize= (20, 6))
plt.title('Comparison of Male and Female Population where Male Population is high', weight= 'bold',
         size= 16)
plt.xticks(size= 15)
plt.show()
```

### Points which we can see in this plot:
    * almost(most of) in all states male population is higher then female population.
    * Kerala is the only state where Female population higher then males.
    * In Chhattisgarh, the male and female population is almost equal(actual difference is 114633). 

```python
# for all the states:

df_com_max.set_index('State / UT').plot.bar(figsize= (20, 6))
plt.title('Comparison of Male and Female Population where Male Population is high', weight= 'bold',
         size= 16)
plt.xticks(size= 15)
plt.show()
```

```python
# checking the difference between Male and Female population

df_MF_diff= df_india_1[['State / UT', 'Males', 'Females']]
df_MF_diff.loc[:, 'Difference in Population(M-F)'] = np.subtract(df_MF_diff['Males'], df_MF_diff['Females'])
```

```python
# top 20 'State / UT' where the difference between male and female population is high:

df_diff_high= df_MF_diff.sort_values(by= 'Difference in Population(M-F)', ascending= False).head(20)
df_diff_high
```

```python
# visualising the above result:

df_diff_high.set_index('State / UT').plot(kind= 'bar', figsize= (10, 6))
plt.show()
```

```python
# top 20 'State / UT' where the difference between male and female population is least:

df_diff_low= df_MF_diff.sort_values(by= 'Difference in Population(M-F)').head(20)
df_diff_low
```

### Point:
    * We can see that 'Kerala' and 'Pondicherry' are the only state where female population is higher then male population.
    * if we calculate, females are almost 6% higher then males

```python
# visualising the above result:

df_diff_low.set_index('State / UT').plot(kind= 'bar', figsize= (10, 6))
plt.show()
```

```python
# visualising only difference:

df_diff_low.set_index('State / UT')['Difference in Population(M-F)'].plot(kind= 'bar', figsize= (10, 6))
plt.xticks(size= 13)
plt.yticks(size= 13)
plt.title('Difference of Male and Female Population(Top 20)', weight= 'bold', size= 14)
plt.show()
```

```python
# visualising the above finding for all the State/ UT

df_MF_diff.set_index('State / UT').sort_values(by= 'Difference in Population(M-F)')['Difference in Population(M-F)'].plot.bar(figsize= (14, 6))
plt.xticks(size= 13)
plt.yticks(size= 13)
plt.show()
```
