```python
# so as we check abve there are 4 rows where every element in Nan. 
# so we can remove them completely

df_car_1= df_car.dropna(axis= 0, how= 'all').reset_index(drop= True)
df_car_1
```

```python
df_car_1.isna().sum()
```

```python
df_car_1[['MSRP', 'Invoice']].isna().sum()
```

```python
# modifying or correcting the data in 'MSRP' and 'Invoice' columns
df_car_1['MSRP']= df_car_1['MSRP'].map(lambda x: x.strip('$').replace(',', '')).astype(float)
df_car_1['Invoice']= df_car_1['Invoice'].map(lambda x: x.strip('$').replace(',', '')).astype(float)
```

```python
# placing the $ sign

df_car_1.rename(columns= {'MSRP' : 'MSRP($)',
                          'Invoice' : 'Invoice($)'}, inplace= True)
```

```python
df_car_1.dtypes
```

```python
df_car_1.isna().sum()
```

```python
# checking the null value rows:

df_car_1[df_car_1['Cylinders'].isna()]
```

```python
df_car_1.iloc[240:255, :]
```

```python
#  filling these missing values with the mean of the column

df_car_1['Cylinders'].fillna(value= round(df_car_1['Cylinders'].mean(), 2), axis= 0, inplace= True)
```

```python
df_car_1.iloc[240:255, :]
```

```python
# checkpoint:

df_car_2= df_car_1.copy()
```

```python
df_car_2.isna().sum()
```

```python
# just selecting few imp results to find some information

df_car_3= df_car_2[['Make', 'Model', 'Type', 'Origin', 'DriveTrain', 'MSRP($)', 'Invoice($)']]
```

```python
df_car_3.head()
```

```python
# lets find the car 'type' according to 'Make'

df_car_3.groupby('Make')['Type'].unique()
```

```python
# number of models per 'Make':

df_car_3.groupby('Make')['Model'].count().sort_values(ascending= False)

# so Toyota has the max numbers of models available(28), after that comes 'Chevrolet(27)', 'Mercedes-Benz(26)', 'Ford(23)', 'BMW(20)', 'Audi(19)'.
# also, 'Hummer' has the least number of models which is only 1.
```

```python
# top 10 "Make" according to model type
df_car_3.groupby('Make')['Model'].count().nlargest(10)
```

```python
# bottom 10 "Make" according to model type
df_car_3.groupby('Make')['Model'].count().nsmallest(10)
```

```python
# Number of type available according to 'Origin' and "Make"

df_car_3.groupby(['Make', 'Origin'])['Type'].nunique()
```

```python
# unique drive train

df_car_3['DriveTrain'].unique()
```

```python
# Drive Train available according to 'Make'

drive_train= df_car_3.groupby('Make')['DriveTrain']
```
