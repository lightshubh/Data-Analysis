```python
# because data is same its just name issue so renaming some columns to fix this issue: 
df_2018.rename(columns= {'State/UT ': 'State/UT', 
                         'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above ' : 'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above - Below 12 Years',
       'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above ' : 'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above - Below 16 Years',
       'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above ' : 'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above - Below 18 Years',
        'Child Victims of Rape (Below 18 Yrs) - Total Girl/Child Victims' : 'Child Victims of Rape (Below 18 Yrs) - Total Girl /Child Victims'},
              inplace= True)
```

```python
# renaming the columns of all 3 datasets:
df_columns= ['S. No', 'Category', 'State/UT', 'Cases Reported',
       'Below 6 Years',
       '6 Years & Above - Below 12 Years',
       '12 Years & Above - Below 16 Years',
       '16 Years & Above - Below 18 Years',
       'Total Girl /Child Victims',
       '18 Years & Above - Below 30 Years',
       '30 Years & Above - Below 45 Years',
       '45 Years & Above - Below 60 Years',
       '60 Years & Above',
       'Total Women/Adult Victims',
       'Total Victims']
```

```python
df_2016.columns = df_2017.columns = df_2018.columns = df_columns
```

```python
for i, j in zip('d', '2016'):
    print(i, j)
```

```python
# removing some columns from all 3 datasets:

datasets= [df_2016, df_2017, df_2018]
for i in datasets:
    i.drop(columns= ['S. No', 'Total Girl /Child Victims', 'Total Women/Adult Victims', 'Total Victims', 'Cases Reported'], inplace= True)
```

```python
df_2016_melted= df_2016.melt(id_vars= ['Category', 'State/UT'], 
             value_vars= ['Below 6 Years', '6 Years & Above - Below 12 Years', '12 Years & Above - Below 16 Years', '16 Years & Above - Below 18 Years', '18 Years & Above - Below 30 Years', '30 Years & Above - Below 45 Years', '45 Years & Above - Below 60 Years', '60 Years & Above'], 
            var_name= 'Age Group', value_name= 'Number of Cases(2016)')
```

```python
df_2017_melted= df_2017.melt(id_vars= ['Category', 'State/UT'], 
             value_vars= ['Below 6 Years', '6 Years & Above - Below 12 Years', '12 Years & Above - Below 16 Years', '16 Years & Above - Below 18 Years', '18 Years & Above - Below 30 Years', '30 Years & Above - Below 45 Years', '45 Years & Above - Below 60 Years', '60 Years & Above'], 
            var_name= 'Age Group', value_name= 'Number of Cases(2017)')
```

```python
df_2018_melted= df_2018.melt(id_vars= ['Category', 'State/UT'], 
             value_vars= ['Below 6 Years', '6 Years & Above - Below 12 Years', '12 Years & Above - Below 16 Years', '16 Years & Above - Below 18 Years', '18 Years & Above - Below 30 Years', '30 Years & Above - Below 45 Years', '45 Years & Above - Below 60 Years', '60 Years & Above'], 
            var_name= 'Age Group', value_name= 'Number of Cases(2018)')
```

```python
# making the table which contain all the total cases column

df_total_table= df_2016_melted.join(df_2017_melted['Number of Cases(2017)']).join(df_2018_melted['Number of Cases(2018)'])
```

```python
df_total_table
```

```python
df_total_table
```

```python
# if we check the result according to state and number of cases in 2016:

df_cat_AGroup= df_total_table.groupby(['Category', 'Age Group'])[['Number of Cases(2016)']].sum()
df_cat_AGroup
```

```python
# if we show these results in plot:

df_cat_AGroup.plot(kind= 'bar', figsize= (14, 6), xlabel= 'State->category', ylabel= 'Number of Cases')
plt.title(label= 'Rape cases according to category(2016)',size= 18)
plt.xticks(size= 13)
plt.yticks(size= 13)
plt.show()

# so maximum victims are from age group "18 Years & Above - Below 30 Years"
```

```python
# checking which state/ut had the maximum number of cases in 2016:

df_state_2016= df_total_table[['Number of Cases(2016)']].groupby(df_total_table['State/UT']).sum()
df_state_2016
```

```python
# visualising the state result:

df_state_2016.plot(kind= 'bar', figsize= (14, 6), ylabel= "Number of Cases")
plt.xlabel(xlabel= 'State/UT', size= 14)
plt.title('Case Reported according to state/ut(2016)', size= 18)
plt.show()

#  so Madhya Pradesh and Uttar Pradesh are in 1st and second place in terms of case reported. 
```

```python
# so what are the number of case reported for MP and UP in 2016:

df_state_2016.sort_values(by= 'Number of Cases(2016)', ascending= False)
```

```python
# cases according to the state and UT
df_total_table.groupby('Category')[['Number of Cases(2016)']].sum()

# so in 2016 the reported cases for state is 32334 and for UT is 1324
```

```python
# lets check the total cases per state/ut from all three years
col= ['Number of Cases(2016)', 'Number of Cases(2017)', 'Number of Cases(2018)']
for i in enumerate(col):
    sns.set_style('whitegrid')
    plt.figure(figsize= (14, 6))
    plt.subplot(2, 2, i[0]+1)
    sns.barplot(x= df_total_table['State/UT'].unique(), y= df_total_table.groupby('State/UT')[i[1]].sum())
    plt.xticks(rotation= 90)
    plt.show()

#plt.xticks(rotation= 90)
```

```python
# lets check the number of cases per age group:
col1= ['Number of Cases(2016)', 'Number of Cases(2017)', 'Number of Cases(2018)']

for i in col1:
    print("Year: %s" %(i[16:20]))
    print('\n')
    print(df_total_table.groupby('Age Group')[[i]].sum())
    print('\n')
```

```python
# lets draw these results:

for i in enumerate(col1):
    plt.figure(figsize= (10, 5))
    plt.subplot(2, 2, i[0]+1)
    sns.barplot(x= df_total_table['Age Group'].unique(), y= df_total_table.groupby('Age Group')[i[1]].sum())
    plt.xticks(rotation= 90)
    plt.title('Cases According to the Age Group(%s)' %(i[1][16:20]))
    plt.show()
```
